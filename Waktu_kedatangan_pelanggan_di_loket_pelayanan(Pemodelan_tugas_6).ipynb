**Tujuan**

Melakukan simulasi untuk memodelkan jumlah kedatangan pelanggan dan waktu antar kedatangan menggunakan distribusi Poisson dan Eksponensial.


**Metodologi**



* Distribusi Poisson digunakan untuk memodelkan jumlah kedatangan pelanggan per unit waktu. Simulasi dilakukan sebanyak 30 iterasi dengan parameter rata-rata λ = 5.
* Distribusi Eksponensial digunakan untuk memodelkan waktu antar kedatangan pelanggan. Simulasi dilakukan sebanyak 30 iterasi dengan parameter rata-rata λ = 5



**Hasil Simulasi Distribusi Poisson**

1. Rata-rata hasil simulasi mendekati nilai teoritis (λ = 5).

2. Hasil simulasi ditampilkan dalam bentuk histogram dan grafik tren jumlah kedatangan pelanggan.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter distribusi Poisson
lambda_poisson = 5  # rata-rata kedatangan per unit waktu
n_iter = 30  # jumlah iterasi simulasi

# Simulasi data Poisson
poisson_samples = np.random.poisson(lambda_poisson, n_iter)

# Menghitung rata-rata hasil simulasi
mean_poisson_simulation = np.mean(poisson_samples)
print(f"Rata-rata hasil simulasi Poisson: {mean_poisson_simulation}")

# Visualisasi histogram hasil simulasi
plt.hist(poisson_samples, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram Hasil Simulasi Distribusi Poisson")
plt.xlabel("Jumlah Kedatangan")
plt.ylabel("Frekuensi")
plt.show()

# Grafik Tren Hasil Simulasi
plt.plot(range(n_iter), poisson_samples, marker='o', color='blue', linestyle='-')
plt.title("Grafik Tren Hasil Simulasi Distribusi Poisson")
plt.xlabel("Iterasi")
plt.ylabel("Jumlah Kedatangan")
plt.show()
```

**Hasil Distribusi Eksponensial**

1. Rata-rata hasil simulasi mendekati nilai teoritis (λ = 5)
2. Hasil simulasi ditampilkan dalam bentuk histogram dan grafik tren waktu antar kedatangan pelanggan.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter distribusi Eksponensial
lambda_exponential = 1 / 5  # rata-rata waktu antar kedatangan (misalnya 5 menit)
n_iter = 30  # jumlah iterasi simulasi

# Simulasi data Eksponensial
exponential_samples = np.random.exponential(1 / lambda_exponential, n_iter)

# Menghitung rata-rata hasil simulasi
mean_exponential_simulation = np.mean(exponential_samples)
print(f"Rata-rata hasil simulasi Eksponensial: {mean_exponential_simulation}")

# Visualisasi histogram hasil simulasi
plt.hist(exponential_samples, bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title("Histogram Hasil Simulasi Distribusi Eksponensial")
plt.xlabel("Waktu Antar Kedatangan (menit)")
plt.ylabel("Frekuensi")
plt.show()

# Grafik Tren Hasil Simulasi
plt.plot(range(n_iter), exponential_samples, marker='o', color='green', linestyle='-')
plt.title("Grafik Tren Hasil Simulasi Distribusi Eksponensial")
plt.xlabel("Iterasi")
plt.ylabel("Waktu Antar Kedatangan (menit)")
plt.show()
```
